```python
df.drop(columns='Original Interest Rate', inplace = True)
```

```python
df.head()
```

```python
df[["Original UPB", 
    "Original Loan Term", 
    'Loan Age', 
    'Original Loan to Value Ratio (LTV)', 
    'Number of Borrowers',
   'Debt-To-Income (DTI)',
   'Minimum Credit Score']] = df[["Original UPB", 
                                  "Original Loan Term", 
                                  'Loan Age', 
                                  'Original Loan to Value Ratio (LTV)', 
                                  'Number of Borrowers',
                                  'Debt-To-Income (DTI)',
                                  'Minimum Credit Score']].astype('int')
```

```python
df.head()
```

```python
# change date columns back to datetime
df['Monthly Reporting Period'] = pd.to_datetime(df['Monthly Reporting Period'])
df['Origination Date'] = pd.to_datetime(df['Origination Date'])
df['Maturity Date'] = pd.to_datetime(df['Maturity Date'])
```

```python
df.head()
```

```python
# mapp dataframe and encode N and Y
df = df.replace(to_replace=['N', 'Y'], value=[0, 1])
```

```python
df.drop(columns='Amortization Type', inplace=True)
```
