```python
# Fill na values for Modification Flag
df['Modification Flag'].fillna('N', inplace=True)
```

```python
df['Modification Flag'].value_counts()
```

```python
# calculate minimum credit score using Borrower and Co-borrower score columns
df['Minimum Credit Score'] = df[['Borrower Credit Score at Origination',
                                 'Co-Borrower Credit Score at Origination']].min(axis=1)
```

```python
for i in df.columns:
    tot = sum(df[i].isnull())
    print(f"{i}: total of {tot} nulls- {round(((tot / len(df)) * 100), 2)}%")
```

```python
df['Foreclosure Date'].value_counts()
```

```python
df['Current Loan Delinquency Status'].value_counts()
```

```python
# Drop Co-borrower and borrower score columns
df.drop(df[['Borrower Credit Score at Origination', 'Co-Borrower Credit Score at Origination']], axis=1, inplace=True)
```

```python
# F T R to be on plan
def check_plan(x):
    if x in ['F', 'T', 'R']:
        return 1
    return 0
```

```python
df['On Assistance Plan'] = df['Borrower Assistance Plan'].apply(lambda x : check_plan(x))
```

```python
df.head()
```

```python
# Dropping Borrower Assistance Plan
df.drop(columns='Borrower Assistance Plan', inplace=True)
```

```python
df['On Assistance Plan'].value_counts()
```

```python
# df['Current Loan Delinquency Status'] >= 4 and not on df['On Assistance Plan'], foreclosed = 1 (True)
# create a list of our conditions
conditions = [
    (df['Current Loan Delinquency Status'] >= 4) &
    (df['On Assistance Plan'] == 0), 
    (df['Foreclosure Date'].notnull())
    ]
# create a list of the values we want to assign for each condition
values = [1,1]
# create a new column and use np.select to assign values to it using our lists as arguments
df['Foreclosed'] = np.select(conditions, values)
```

```python
df.head()
```

```python
df.drop(df[['Current Loan Delinquency Status', 'On Assistance Plan', 'Foreclosure Date']], axis=1, inplace=True)
```

```python
df[['Monthly Reporting Period',
        'Origination Date', 
        'Maturity Date']]  = df[['Monthly Reporting Period',
                                    'Origination Date', 
                                    'Maturity Date']].astype('str')
```

```python
df['Monthly Reporting Period'] = pd.to_datetime(df['Monthly Reporting Period'],format='%m%Y')
```

```python
 # new data frame with split value columns
df['Origination Date'] = df['Origination Date'].str.split(".", n = 1, expand = True)
```

```python
df['Origination Date'] = pd.to_datetime(df['Origination Date'],format='%m%Y')
```

```python
 # new data frame with split value columns
df['Maturity Date'] = df['Maturity Date'].str.split(".", n = 1, expand = True)
df['Maturity Date'] = pd.to_datetime(df['Maturity Date'], format='%m%Y')
df.head()
```
